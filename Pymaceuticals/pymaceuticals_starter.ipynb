```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
study_data_raw = pd.merge(study_results,mouse_metadata, on=["Mouse ID", "Mouse ID"], how = 'left')

#study_data_list = study_data_raw[(study_data_raw["Timepoint"] == 0)]
#study_data_list.reset_index()

# Display the data table for preview
study_data_raw.head()
```

```python
# Checking the number of mice.
#study_data_list["Mouse ID"].count()
study_data_raw["Mouse ID"].nunique()
#len(study_data_list["Mouse ID"].unique())
#study_data_list["Mouse ID"].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate = study_data_raw[study_data_raw.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)]
duplicate_Mouse_ID = duplicate.iloc[0,0]
#print(duplicate)
duplicate_Mouse_ID
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_df = study_data_raw.loc[study_data_raw["Mouse ID"] == duplicate_Mouse_ID]
duplicate_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
study_data_clean = study_data_raw.drop_duplicates("Timepoint")
study_data_clean = study_data_raw[study_data_raw['Mouse ID']!="g989"]
study_data_clean
```

```python
# Checking the number of mice in the clean DataFrame.
study_data_clean["Mouse ID"].nunique()
```

## Summary Statistics

```python
drug_data = study_data_clean.groupby(["Drug Regimen"])
#-----(don't type 5 in .head()) wierd stuff happens ---> shows full list
drug_data.head(10)
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen


# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 


#Note to future self, Drug Regimen needed to be sorted first then sorted by tumor volume for less stress


Mean_Tumor_Volume = drug_data["Tumor Volume (mm3)"].mean()
Median_Tumor_Volume = drug_data["Tumor Volume (mm3)"].median()
Tumor_Volume_Variance = drug_data["Tumor Volume (mm3)"].var()
Tumor_Volume_Std_Dev = drug_data["Tumor Volume (mm3)"].std()
Tumor_Volume_Std_Err = drug_data["Tumor Volume (mm3)"].sem()


# Assemble the resulting series into a single summary DataFrame.

tumor_data_df = pd.DataFrame({
    "Mean Tumor Volume":Median_Tumor_Volume,
    "Median Tumor Volume":Median_Tumor_Volume,
    "Tumor Volume Variance":Tumor_Volume_Variance,
    "Tumor Volume Std. Dev.":Tumor_Volume_Std_Dev,
    "Tumor Volume Std. Err.":Tumor_Volume_Std_Err})

tumor_data_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.
#.agg()
agg_drug_tumor_summary = study_data_clean.groupby("Drug Regimen")[["Tumor Volume (mm3)"]].agg(['mean','median','var','std','sem'])
agg_drug_tumor_summary
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

#for scatter plot
x_axis = mouse_id_avg_list["Weight (g)"]
y_axis = mouse_id_avg_list["Tumor Volume (mm3)"]


#regression line 
corr=st.pearsonr(x_axis,y_axis)[0]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_axis, y_axis)
regress_values = x_axis * slope + intercept

#for the display
plt.scatter(x_axis, y_axis)
plt.plot(x_axis, regress_values, "r-")
plt.title("Average Tumor Volume vs. Mouse Weight for the Capomulin Regimen")
plt.xlabel("Average Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
print(f"The correlation between mouse weight and average tumor volume is: {corr:.2f}")
plt.show()
```
